```python
Bank_df_1 = pd.DataFrame({'Client ID':[1,2,3,4,5],
                          'First Name':['Ahtsham','Ali','Imran','Rizwan','Haad'],
                          'Last Name':['Shah', 'Hassan', 'Hussain', 'Ali', 'Kazmi'],
                          }, index=[0,1,2,3,4])
Bank_df_1
```

```python
Bank_df_2 = pd.DataFrame({'Client ID':[6,7,8,9,10],
                          'First Name':['Kamran','Fazeel','Aqib','Awais','Junaid'],
                          'Last Name':['Haider', 'Sial', 'Qureshi', 'Khan', 'Raja'],
                          }, index=[5,6,7,8,9])

Bank_df_2
```
